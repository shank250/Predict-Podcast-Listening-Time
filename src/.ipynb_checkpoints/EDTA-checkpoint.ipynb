## Analysing the relationship between Episode_Title and Listening_Time_minutes

### TF-IDF (Term Frequency–Inverse Document Frequency)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
import numpy as np

# TF-IDF vectorization
tfidf = TfidfVectorizer(max_features=500)  # Limit to 500 terms for performance
X_title = tfidf.fit_transform(train['Episode_Title'])

# Model to predict y using only title
model = Ridge()
scores = cross_val_score(model, X_title, train['Listening_Time_minutes'], cv=5, scoring='neg_root_mean_squared_error')
print("RMSE using only Episode_Title:", -np.mean(scores))
```

```python
import pandas as pd
import re

# Assuming your dataframe is named df
def extract_episode_number(title):
    match = re.search(r'\bEpisode\s+(\d+)\b', title, flags=re.IGNORECASE)
    return int(match.group(1)) if match else None

# Create the new column
train['Episode_Number'] = train['Episode_Title'].apply(extract_episode_number)
```

```python
train.head()
```

```python
import matplotlib.pyplot as plt

plt.hexbin(train['Episode_Number'], train['Listening_Time_minutes'], gridsize=50, cmap='Blues')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data (assuming already loaded)
# df = pd.read_csv('train.csv')  # example

# Optional: drop NaNs to avoid errors in KDE
df_kde = train[['Episode_Length_minutes', 'Listening_Time_minutes']].dropna()

# Downsample if dataset is very large
df_kde = df_kde.sample(n=100000, random_state=42)  # tune this if needed

# KDE Plot
sns.kdeplot(
    data=df_kde,
    x='Episode_Length_minutes',
    y='Listening_Time_minutes',
    cmap='mako',
    fill=True,
    thresh=0.05
)

plt.title('Density of Episode Length vs Listening Time')
plt.xlabel('Episode Length (minutes)')
plt.ylabel('Listening Time (minutes)')
plt.show()
```

Thus we can say that both the episode length and the listening time both are a bit related.
